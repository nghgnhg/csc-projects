## merge df

```python
reviews.shape,reviews.columns
```

```python
products.shape,products.columns
```

```python
products['product_id'] = products['item_id']
```

```python
data = reviews.merge(products,how='left',on='product_id')
data.drop('item_id',axis=1,inplace=True)

print(data.shape)
data.head(2)
```

```python
data.columns
```

```python
df = data[['customer_id','product_id','rating_x','name','list_price','brand','group']]
df.columns = ['customer','product','rating','name','price','brand','group']
df.head()
```

# ----------
### Products - brand

```python
df_brand = products.groupby('brand').agg({
    'rating':['mean','std'],
    'price':['mean','std'],
    'list_price':['mean','std','count']
}).round(2)
df_brand.columns = ['rating_mean','rating_std','price_mean','price_std',
                    'listprice_mean','listprice_std','items']

df_brand['%count'] = round((df_brand['items']/df_brand['items'].sum())*100, 2)
df_brand
```

```python
df_brand.sort_values(['items','rating_mean'],ascending=False)#.sort_values('rating_mean',ascending=False)
```

```python
# 5 nhãn hàng có giá trung bình sản phẩm cao nhất
df_brand.sort_values('price_mean',ascending=False)[['price_mean','items']].head()
```

```python
# 5 nhãn hàng có giá trung bình sản phẩm thấp nhất
df_brand.sort_values('price_mean')[['price_mean','items']].head()
```

```python
# 5 nhãn hàng có nhiều sản phẩm bán trên tiki nhất
df_brand.sort_values('items',ascending=False)[['items','%count']].head()
```

### Reviews - brand

```python
df.columns
```

```python
rv_brand = df.groupby('brand').agg({
    'rating':['mean','std'],
    'price':['mean','std','count']
}).round(2)
rv_brand.columns = ['rating_mean','rating_std','price_mean','price_std','customer']

rv_brand['%customer'] = round((rv_brand['customer']/rv_brand['customer'].sum())\
                              *100, 2)
```

```python
# 5 thương hiệu được khách hàng đánh giá nhiều nhất
rv_brand.sort_values('customer',ascending=False).head()
```

```python
# 5 thương hiệu có giá trung bình được khách hàng chi tiền nhiều nhất (trung bình 1 khách hàng)
rv_brand.sort_values('price_mean',ascending=False).head()
```

### Products - group

```python
df_group = products.groupby('group').agg({
    'rating':['mean','std'],
    'price':['mean','std'],
    'list_price':['mean','std','count']
}).round(2)
df_group.columns = ['rating_mean','rating_std','price_mean','price_std',
                    'listprice_mean','listprice_std','items']

df_group['%count'] = round((df_group['items']/df_group['items'].sum())*100, 2)
df_group
```

```python
# 5 nhóm sản phẩm được bán trên tiki nhiều nhất
df_group.sort_values('items',ascending=False)[['items','%count']].head()
```

```python
# 5 nhóm sản phẩm có giá trung bình cao nhất
df_group.sort_values('price_mean',ascending=False)[['price_mean','items']].head()
```

```python
# 5 nhóm sản phẩm có giá trung bình thấp nhất
df_group.sort_values('price_mean')[['price_mean','items']].head()
```

### Reviews - group

```python
rv_group = df.groupby('group').agg({
    'rating':['mean','std'],
    'price':['mean','std','count']
}).round(2)
rv_group.columns = ['rating_mean','rating_std','price_mean','price_std','customer']

rv_group['%customer'] = round((rv_group['customer']/rv_group['customer'].sum())\
                              *100, 2)
```

```python
# 5 nhóm sản phẩm có khách hàng đánh giá nhiều nhất
rv_group.sort_values('customer',ascending=False).head()
```

```python
# 5 nhóm sản phẩm có khách hàng đánh giá cao nhất
rv_group.sort_values('rating_mean',ascending=False)[['rating_mean','customer','%customer']].head()
```

```python
# 5 nhóm sản phẩm có khách hàng đánh giá thấp nhất
rv_group.sort_values('rating_mean')[['rating_mean','customer','%customer']].head()
```
